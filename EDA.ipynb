```python
plt.figure(figsize=(8, 5))
sns.histplot(df['vomitoxin_ppb'], bins=30, kde=True)
plt.title('Distribution of DON Concentration')
plt.show()
```

```python
# Boxplot to Identify Outliers
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['vomitoxin_ppb'])
plt.title('Boxplot of DON Concentration')
plt.show()
```

```python
df['DON_Concentration'] = np.log1p(df['vomitoxin_ppb'])
```

```python
Q1 = df['DON_Concentration'].quantile(0.25)
Q3 = df['DON_Concentration'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier threshold
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Keep only non-outliers
df_cleaned = df[(df['DON_Concentration'] >= lower_bound) & (df['DON_Concentration'] <= upper_bound)]

# Save the cleaned dataset
df_cleaned.to_csv('data/cleaned_data.csv', index=False)
print(f"Shape after removing outliers: {df_cleaned.shape}")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned['DON_Concentration'])
plt.title('Boxplot of DON Concentration (After Outlier Removal)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['DON_Concentration'], bins=30, kde=True)
plt.title('Distribution of DON Concentration (After Log Transform & Outlier Removal)')
plt.xlabel('DON Concentration')
plt.ylabel('Frequency')
plt.show()
```
